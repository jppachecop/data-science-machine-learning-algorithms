```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
%matplotlib inline

#Criação de um dataset com 3 clusters e um desvio padrão de 0,5
#para cada cluster.
#Número de amostras de 400 e 3 pontos como centróides

#Determinação dos centróides
centers = [[0.5, 2], [-1, -1], [1.5, -1]]

#Criação do dataset
X, y = make_blobs(n_samples=400, centers=centers, 
                  cluster_std=0.5, random_state=0)

#Normalização dos valores
X = StandardScaler().fit_transform(X)

#Gráfico do dataset para ver cada cluster
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='Paired')
```

```python
#Objeto DBSCAN e ajuste dos dados
db = DBSCAN(eps=0.4, min_samples=20)
db.fit(X)


#Determinamos os valores de eps e minPts
#através dos parâmetros eps e min_samples
y_pred = db.fit_predict(X)

plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1],c=y_pred, cmap='Paired')
plt.title("Clusters determined by DBSCAN")

#Aqui temos o número de outliers
db.labels_[db.labels_ == -1].size
```
